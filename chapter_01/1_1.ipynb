## 1.1.1 선형대수학 소개

* 방정식은 등호를 기준으로 좌우에 있는 2개의 수학적 표현이 동일함을 나타낸다.
* 일차방정식은 x, y, z와 같은 변수에 대하여 모든 변수의 지수(거듭제곱)가 1 또는 0인 방정식이다.

* 삼각함수의 편각(argument), 지수함수, 로그함수 또는 쌍곡선함수를 포함하는 방정식은 일차방정식이 아니다.
* 일차방정식의 집합을 연립일차방정식(linear system)이라고 한다.
* 연립일차  방정식에 관한 질문
1. 어떤 해가 존재할까?
2. 연립일차방정식은 해가 없거나, 해가 유일하거나, 해가 무수히 많은 경우로 나뉘는가?
3. 해가 존재한다면, 어떻게 모든 해를 찾을 수 있는가?
4. 해는 어떤 구조로 이루어져 있는가?

* 선형대수학은 일차방정식을 체계적으로 탐구하는 학문이다.
* 그러나 행렬과 벡터는 내적공간을 필요로 하는 함수해석학과 양자역학 분야에 다양하게 응용된다.
* 또 다른 응용 분야로는 최적화(optimization)과 근사(approximation)가 있다.
* 선형 문제는 이해하기 쉬운 데다 많은 정보를 알려주기 때문에 우리는 수학의 많은 영역을 선형 문제로 전환하여 해결하려고 한다.

## 1.1.2 연립일차방정식


```python
import numpy as np
import matplotlib.pyplot as plt

# 첫 번째 일차함수의 계수와 상수
a1 = -2  # 기울기
b1 = 2.5  # y절편

# 두 번째 일차함수의 계수와 상수
a2 = -1  # 기울기
b2 = 3/2  # y절편

# x값의 범위 설정
x = np.linspace(-10, 10, 100)  # -10부터 10까지 100개의 점을 생성

# 각 일차함수의 y값 계산
y1 = a1 * x + b1
y2 = a2 * x + b2

# 그래프 그리기
plt.plot(x, y1, label='first')
plt.plot(x, y2, label='second')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 일차방정식의 계수와 상수
a = 1  # x에 곱해지는 계수
b = 1  # y에 곱해지는 계수
c = 2  # z에 곱해지는 계수
d = 0  # 상수항

# 3차원 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x, y 값 범위 설정
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

# x, y 값 조합으로 z 값 계산
X, Y = np.meshgrid(x, y)
Z = a*X + b*Y + c

# 일차방정식 그래프 그리기
ax.plot_surface(X, Y, Z, alpha=0.5)

# 축 레이블 설정
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 그래프 보여주기
plt.show()
```

* 일반적으로 유한개의 미지수 x,y,z,w, ...로 이루어진 유한개의 일차방정식을 연립일차방정식(system of linear equations 또는 linear system)이라고 한다.

* 연립일차방정식에 다음과 같은 작업을 수행할 수 있다.
1. 두 방정식의 위치를 바꾼다.
2. 방정식에 0이 아닌 상수를 곱한다.
3. 한 방정식에서 다른 방정식을 빼거나 더한다.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 일차방정식의 계수와 상수
a1, b1, c1, d1 = 1, 2, 4, 7 
a2, b2, c2, d2 = 3, 7, 2, -11
a3, b3, c3, d3 = 2, 3, 3, 1

# 3차원 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x, y 값 범위 설정
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

# x, y 값 조합으로 z 값 계산
X1, Y1 = np.meshgrid(x, y)
Z1 = a1*X + b1*Y + c1 + d

X2, Y2 = np.meshgrid(x, y)
Z2 = a2*X + b2*Y + c2 + d

X3, Y3 = np.meshgrid(x, y)
Z3 = a3*X + b3*Y + c3 + d

# 일차방정식 그래프 그리기
ax.plot_surface(X1, Y1, Z1, alpha=0.5)
ax.plot_surface(X2, Y2, Z2, alpha=0.5)
ax.plot_surface(X3, Y3, Z3, alpha=0.5)

# 축 레이블 설정
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 그래프 보여주기
plt.show()
```

* 세 평면의 교점이 바로 연립일차방정식의 해이다.

## 1.1.3 해의 유형

### 해가 없는 경우

```python
import numpy as np
import matplotlib.pyplot as plt

# 첫 번째 일차함수의 계수와 상수
a1 = -2/3  # 기울기
b1 = 2  # y절편

# 두 번째 일차함수의 계수와 상수
a2 = -4/6  # 기울기
b2 = 9/6  # y절편

# x값의 범위 설정
x = np.linspace(-10, 10, 100)  # -10부터 10까지 100개의 점을 생성

# 각 일차함수의 y값 계산
y1 = a1 * x + b1
y2 = a2 * x + b2

# 그래프 그리기
plt.plot(x, y1, label='first')
plt.plot(x, y2, label='second')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function')
plt.grid(True)
plt.legend()
plt.show()
```

* 두 직선은 평행하므로 교점이 없다.
* 따라서 주어진 연립일차방정식의 해는 없다.
* 해가 없는 연립일차방정식을 해를 갖지 않는(inconsistent) 연립일차방정식이라고 한다.
* 반대로 연립일차방정식의 해가 적어도 하나 이상이면 이 연립일차방정식을 해를 갖는(consistent) 연립일차방정식이라고 한다.

## 해가 무수히 많은 경우

```python
import numpy as np
import matplotlib.pyplot as plt

# 첫 번째 일차함수의 계수와 상수
a1 = -4/6  # 기울기
b1 = 12/6  # y절편

# 두 번째 일차함수의 계수와 상수
a2 = -2/3  # 기울기
b2 = 6/3  # y절편

# x값의 범위 설정
x = np.linspace(-10, 10, 100)  # -10부터 10까지 100개의 점을 생성

# 각 일차함수의 y값 계산
y1 = a1 * x + b1
y2 = a2 * x + b2

# 그래프 그리기
plt.plot(x, y1, label='first')
plt.plot(x, y2, label='second')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function')
plt.grid(True)
plt.legend()
plt.show()
```

### 핵심 요약
* 일차방정식은 미지수의 지수가 1 또는 0인 방정식이다.
* 연립일차방정식은 소거법을 이용하여 푼다.
