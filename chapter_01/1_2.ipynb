## 1.2.2 기본 행 연산

* 첨가행렬(argumented matrix) - 미지수의 계수와 우변의 상수를 포함하는 행렬
* 'argumented'는 '늘린(to increase)'이라는 의미이다.
* 행렬을 늘린다는 것은 기존 행렬에 하나 이상의 열을 첨가한다는 뜻이다.
* 첨가행렬의 연산 수행 -> 기본 행 연산(elementary row operation)
1. 0이 아닌 상수를 한 행에 곱한다.
2. 1번에서 얻은 행을 다른 행에 더하거나 뺀다.
3. 행을 교환(interchange rows)한다.


## 1.2.3 가우스 소거법

* 가우스 소거법(Gaussian elimination) - 미지수 x, y,z, w, ...의 값을 찾기 위해 기본 행 연산 1,2,3을 수행하여 연립일차방정식의 해를 구하는 방법

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 일차방정식의 계수와 상수
a1, b1, c1, d1 = 1, -3, 5, -9
a2, b2, c2, d2 = 2, -1, -3, 19
a3, b3, c3, d3 = 3, 1, 4, -13

# 3차원 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x, y 값 범위 설정
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

# x, y 값 조합으로 z 값 계산
X1, Y1 = np.meshgrid(x, y)
Z1 = a1*X1 + b1*Y1 + c1 + d1

X2, Y2 = np.meshgrid(x, y)
Z2 = a2*X2 + b2*Y2 + c2 + d2

X3, Y3 = np.meshgrid(x, y)
Z3 = a3*X3 + b3*Y3 + c3 + d3

# 일차방정식 그래프 그리기
ax.plot_surface(X1, Y1, Z1, alpha=0.5)
ax.plot_surface(X2, Y2, Z2, alpha=0.5)
ax.plot_surface(X3, Y3, Z3, alpha=0.5)

# 축 레이블 설정
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 그래프 보여주기
plt.show()
```

* 후방 대입(back substitution)을 이용한 가우스 소거법
* 가우스 소거법의 목적은 행렬의 좌측 하단 모서리를 0으로 바꾸어 '삼각'행렬을 만드는 것이다.
* 삼각행렬은 다음과 같은 기본 행 연산을 통해 얻을 수 있다.
1. 0이 아닌 상수를 한 행에 곱한다.
2. 1번 과정에서 얻은 행을 다른 행에 더하거나 뺀다.
3. 행을 교환한다.
* 행동치(row equivalent) - 위 세 가지 연산을 이용하여 한 행렬이 다른 행렬로 변환되는 경우 두 행렬을 의미


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 일차방정식의 계수와 상수
a1, b1, c1, d1 = 1, 3, 2, 13
a2, b2, c2, d2 = 4, 4, -3, 3
a3, b3, c3, d3 = 5, 1, 2, 13

# 3차원 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x, y 값 범위 설정
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

# x, y 값 조합으로 z 값 계산
X1, Y1 = np.meshgrid(x, y)
Z1 = a1*X1 + b1*Y1 + c1 + d1

X2, Y2 = np.meshgrid(x, y)
Z2 = a2*X2 + b2*Y2 + c2 + d2

X3, Y3 = np.meshgrid(x, y)
Z3 = a3*X3 + b3*Y3 + c3 + d3

# 일차방정식 그래프 그리기
ax.plot_surface(X1, Y1, Z1, alpha=0.5)
ax.plot_surface(X2, Y2, Z2, alpha=0.5)
ax.plot_surface(X3, Y3, Z3, alpha=0.5)

# 축 레이블 설정
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 그래프 보여주기
plt.show()
```

## 1.2.4 행 연산의 연장

* 기약행 사다리꼴(reduced row echelon form, rref)의 네 가지 조건
1. 모든 성분이 0인 행은 행렬의 마지막 행으로 둔다.
2. 0이 아닌 성분이 있는 행에서 첫 번째로 0이 아닌 성분은 1이다. 여기서, 1을 선행 1(leading 1)이라고 한다.
3. 모든 성분이 0이 아닌 인접한 두 행의 선행 1은 행렬의 왼쪽 상단에서 오른쪽 하단으로 단계적으로 구성된다.
4. 선행 1을 포함하는 열에서 0이 아닌 유일한 성분은 선행 1이다.
* 조건 4만 만족하지 않는 행렬은 행 사다리꼴(row echelon form)이라고 한다.
* 선행 1은 선행계수(leading coefficient)라고 부르기도 한다.
* 첨가행렬을 행 사다리꼴로 변환하는 과정을 가우스 소거법이라고 한다.
* 첨가행렬을 기약행 사다리꼴로 변환하는 과정을 가우스-조단 소거법(Gauss-Jordan elimination)이라고 한다.
